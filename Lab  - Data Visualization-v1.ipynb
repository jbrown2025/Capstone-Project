**Install and Import Necessary Python Libraries**

Ensure that you have the required libraries installed to work with SQLite and Pandas:


```python
import pandas as pd
import matplotlib.pyplot as plt
```

**Read the CSV File into a Pandas DataFrame**

Load the Stack Overflow survey data into a Pandas DataFrame:


```python
# Read the CSV file
df = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv')

# Display the first few rows of the data
df.head()
```

**Create a SQLite Database and Insert the Data**

Now, let's create a new SQLite database (`survey-data.sqlite`) and insert the data from the DataFrame into a table using the sqlite3 library:


```python
import sqlite3

# Create a connection to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Write the dataframe to the SQLite database
df.to_sql('main', conn, if_exists='replace', index=False)


# Close the connection
conn.close()
```

**Verify the Data in the SQLite Database**
Verify that the data has been correctly inserted into the SQLite database by running a simple query:


```python
# Reconnect to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Run a simple query to check the data
QUERY = "SELECT * FROM main LIMIT 5"
df_check = pd.read_sql_query(QUERY, conn)

# Display the results
print(df_check)
```

## Demo: Running an SQL Query


Count the number of rows in the table named 'main'


```python
QUERY = """
SELECT COUNT(*) 
FROM main
"""
df = pd.read_sql_query(QUERY, conn)
df.head()
```

## Demo: Listing All Tables


To view the names of all tables in the database:


```python
QUERY = """
SELECT name as Table_Name FROM sqlite_master 
WHERE type = 'table'
"""
pd.read_sql_query(QUERY, conn)
```

## Demo: Running a Group By Query
    
For example, you can group data by a specific column, like Age, to get the count of respondents in each age group:


```python
QUERY = """
SELECT Age, COUNT(*) as count
FROM main
GROUP BY Age
ORDER BY Age
"""
pd.read_sql_query(QUERY, conn)
```

## Demo: Describing a table

Use this query to get the schema of a specific table, main in this case:


```python
table_name = 'main'

QUERY = """
SELECT sql FROM sqlite_master 
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY, conn)
print(df.iat[0,0])
```

## Hands-on Lab


### Visualizing the Distribution of Data

**Histograms**

Plot a histogram of CompTotal (Total Compensation).


```python
## Write your code here

QUERY = """
SELECT CompTotal
FROM main
GROUP BY CompTotal
"""
distro = pd.read_sql_query(QUERY, conn)

plt.figure(figsize=(12,6))
plt.hist(distro, bins = 100)
plt.title('Distribution of Total Compensation')
plt.xlabel('Total Compensation')
plt.xticks(rotation = 90)
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

**Box Plots**

Plot a box plot of Age.


```python
## Write your code here
import seaborn as sns
QUERY = """
SELECT Age, COUNT(*) as count
FROM main
GROUP BY Age
"""
df = pd.read_sql_query(QUERY, conn)

plt.figure(figsize= (12,6))
sns.boxplot(x=df['Age'])
plt.title('Box plot of Age')
plt.xticks(rotation=45)
plt.show()
```

### Visualizing Relationships in Data

**Scatter Plots**

Create a scatter plot of Age and WorkExp.


```python
## Write your code here
QUERY = """
SELECT Age, WorkExp
FROM main
GROUP BY WorkExp
"""
df = pd.read_sql_query(QUERY, conn)

plt.figure(figsize=(14, 8))
sns.scatterplot(data=df, x='Age', y='WorkExp')
plt.title('Age vs Work Experience')
plt.xlabel('Age')
plt.ylabel('Years Experience')
plt.show()
```

**Bubble Plots**

Create a bubble plot of `TimeSearching` and `Frustration` using the Age column as the bubble size.


```python
## Write your code here
QUERY = """
SELECT TimeSearching, Frustration, Age
FROM main
GROUP BY TimeSearching
"""
df = pd.read_sql_query(QUERY, conn)

plt.figure(figsize=(14, 8))
sns.scatterplot(data=df, x='TimeSearching', y='Frustration', hue = 'Age', size= 'Age')
plt.title('Bubble Plot of Timesearching, Frustration by age')
plt.xlabel('Time Searching')
plt.ylabel('Frustration')
plt.show()
```

### Visualizing Composition of Data

**Pie Charts**

Create a pie chart of the top 5 databases(`DatabaseWantToWorkWith`) that respondents wish to learn next year.


```python
## Write your code here
QUERY = """
SELECT DatabaseWantToWorkWith, COUNT(*) as count
FROM main
GROUP BY DatabaseWantToWorkWith
ORDER BY DatabaseWantToWorkWith DESC LIMIT 5
"""
df = pd.read_sql_query(QUERY, conn)

explode_list = [0, 0, 0.1, 0.1, 0] 

plt.figure(figsize=(10, 6))
plt.pie(df['count'], labels = None, autopct='%1.1f%%', explode=explode_list,
        startangle=0, pctdistance=1.12)
plt.legend(labels=df['DatabaseWantToWorkWith'], loc='upper left', fontsize=10) 
plt.title('Distribution of Top Databases Survey wants to work with')
plt.axis('equal')
plt.show()
```

**Stacked Charts** 

Create a stacked bar chart of median `TimeSearching` and `TimeAnswering` for the age group 30 to 35.


```python
## Write your code here
QUERY = """
SELECT TimeSearching, TimeAnswering
FROM main
WHERE AGE == "25-34 years old"
"""
df = pd.read_sql_query(QUERY, conn)

timemap = {None: 0, '15-30 minutes a day': 22.5, '30-60 minutes a day':45, '60-120 minutes a day':90,
           'Less than 15 minutes a day': 7.5, 'Over 120 minutes a day': 150}
df['TimeSearching'] = df['TimeSearching'].map(timemap)
df['TimeAnswering'] = df['TimeAnswering'].map(timemap)

medians = df.median()
df1 = medians.reset_index()
df1.columns = ['Time', 'Median']

plt.figure(figsize=(12,6))
sns.barplot(data = df1, x = 'Time', y='Median', alpha = 0.5)
plt.title('Median Times Answering and Searching for 30-35 Year Olds')
plt.xlabel('Time')
plt.ylabel('Years')

plt.show()
```

### Visualizing Comparison of Data

**Line Chart**

Plot the median `CompTotal` for all ages from 45 to 60.


```python
## Write your code here
QUERY = """
SELECT CompTotal
FROM main
WHERE AGE IN ("45-54 years old", "55-64 years old")
"""
df = pd.read_sql_query(QUERY, conn)

medians = df['CompTotal'].median()
df1 = pd.DataFrame({'Compensation': ['CompTotal'], 'Median': [medians]})

plt.figure(figsize=(12,6))
sns.barplot(data = df1, x = 'Compensation', y='Median', alpha = 0.5)
plt.title('Median Compensation for 45-60 Year Olds')
plt.xlabel('45-60 Year Olds')
plt.ylabel('Compensation')

plt.show()
```

**Bar Chart**

Create a horizontal bar chart using the `MainBranch` column.


```python
## Write your code here
QUERY = """
SELECT MainBranch, COUNT(*) as count
FROM main
GROUP BY MainBranch
"""
df = pd.read_sql_query(QUERY, conn)

plt.figure(figsize=(12,6))
sns.barplot(data = df, y = 'MainBranch', x='count', alpha = 0.5)
plt.title('Distribution of Main Branch')
plt.xlabel('Count')
plt.ylabel('Main Branch')

plt.show()
```
